Part 1: Visualizing the First Two Features of the Iris Dataset

```python
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
data = iris.data
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=iris.target)
plt.colorbar()
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Original Iris Data (First Two Features)')
plt.show()
```

Part 2: Autoencoder for Dimensionality Reduction and Visualization

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
```

```python
data -= np.mean(data, axis=0)
data /= np.std(data, axis=0)
```

```python
encoding_dim = 2  # 2D representation
input_data = Input(shape=(4,))
encoded = Dense(encoding_dim, activation='relu')(input_data)
decoded = Dense(4, activation='sigmoid')(encoded)
autoencoder = Model(input_data, decoded)
```

```python
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
autoencoder.fit(data, data, epochs=100, batch_size=10, shuffle=True)
```

```python
encoder = Model(input_data, encoded)
encoded_data = encoder.predict(data)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(encoded_data[:, 0], encoded_data[:, 1], c=iris.target)
plt.colorbar()
plt.xlabel('Encoded Feature 1')
plt.ylabel('Encoded Feature 2')
plt.title('2D Visualization of Encoded Iris Data')
plt.show()
```
